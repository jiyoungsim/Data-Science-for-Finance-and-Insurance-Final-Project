```python
# import modules
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.svm import SVC
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Problem 7

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics.regression import r2_score

# import data
data = pd.read_csv('HW4-Data1.csv', usecols=['X','Y'])
X = data['X']
y = data['Y']
```

```python
# Problem 7(a)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
plt.figure()
plt.scatter(X_train, y_train, label='training data')
plt.scatter(X_test, y_test, label='test data')
plt.legend(loc=4);
```

```python
# Problem 7(b)

def answer_7b():
    result = np.zeros((4,100))
    
    for i, degree in enumerate([1,3,6,9]):
        poly = PolynomialFeatures(degree=degree)
        X_poly = poly.fit_transform(X_train.values.reshape(11,1))
        linreg = LinearRegression().fit(X_poly, y_train)
        y = linreg.predict(poly.fit_transform(np.linspace(0,10,100).reshape(100,1)));
        result[i,:] = y
    return result
```

```python
def plot_7b(degree_predictions):
    import matplotlib.pyplot as plt
    get_ipython().magic('matplotlib notebook')
    plt.figure(figsize=(10,5))
    plt.plot(X_train, y_train, 'o', label='training data', markersize=10)
    plt.plot(X_test, y_test, 'o', label='test data', markersize=10)
    for i,degree in enumerate([1,3,6,9]):
        plt.plot(np.linspace(0,10,100), degree_predictions[i], alpha=0.8, lw=2, label='degree={}'.format(degree))
    plt.ylim(-1,2.5)
    plt.legend(loc=4)

plot_7b(answer_7b())
```

```python
def answer_7c():
    r2_train = np.zeros(10)
    r2_test = np.zeros(10)
    
    for deg in range(10):
        poly = PolynomialFeatures(degree=deg)
        
        # Train and score x_train
        X_poly = poly.fit_transform(X_train.values.reshape(11,1))
        clf = LinearRegression().fit(X_poly, y_train)        
        r2_train[deg] = clf.score(X_poly, y_train);
        
        # Score x_test (do not train)
        X_test_poly = poly.fit_transform(X_test.values.reshape(4,1))
        r2_test[deg] = clf.score(X_test_poly, y_test)
        
    return r2_train, r2_test

r2_train, r2_test = answer_7c()

degrees = np.arange(0, 10)
plt.figure()
plt.plot(degrees, r2_train,label = 'Train')
plt.plot(degrees, r2_test, label = 'Test')
plt.legend()

print('Degree 0 corresponds to the model that is the most underfitting. Degree 9 corresponds to the overfitting model. Degree 6 would provide a model with good generalization performance on this dataset.')
```

```python
def answer_7d():
    from sklearn.linear_model import Lasso
    # Create Polinomial Features
    poly = PolynomialFeatures(degree=12)
    
    # Shape Polinomial Features
    X_train_poly = poly.fit_transform(X_train.values.reshape(11,1))
    X_test_poly = poly.fit_transform(X_test.values.reshape(4,1))
    
    # Linear Regression
    linreg = LinearRegression().fit(X_train_poly, y_train)
    lin_r2_test = linreg.score(X_test_poly, y_test)

    # Lasso Regression
    linlasso = Lasso(alpha=0.01, max_iter = 10000).fit(X_train_poly, y_train)
    las_r2_test = linlasso.score(X_test_poly, y_test)
    
    return lin_r2_test, las_r2_test

answer_7d()
```

```python
# Problem 8

# Create classification dataset
from sklearn.datasets import make_blobs 
X_2, y_2 = make_blobs(n_samples = 100, n_features = 2, centers = 8, cluster_std = 1.3, random_state = 4) 
y_2 = y_2 % 2 
```

```python
# Problem 8(a)

X_train, X_test, y_train, y_test = train_test_split(X_2, y_2, random_state=0)
```

```python
plt.figure()
plt.scatter(X_train[:, 0], X_train[:, 1], marker='o', c = y_train, label = 'Training Data')
plt.scatter(X_test[:, 0], X_test[:, 1], marker='x', c = y_test, label = 'Test Data')
plt.legend(loc = 3);
```

```python
# Problem 8(b)

def make_meshgrid(x, y, h=.02):
    # Create a mesh of points to plot in
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy

def find_Z(clf, xx, yy):
    # Prediction
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Decision function for each datapoint on the grid
    Z1 = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z1 = Z1.reshape(xx.shape)
    
    return Z, Z1

def answer_8b(X0, X1):
    from matplotlib.colors import ListedColormap

    cmap_light = ListedColormap(['#FFFF99', '#00FFCC', '#0000FF','#BCBCBC'])
    cmap_bold = ListedColormap(['#FFFF00', '#00FF00', '#0000FF','#000000'])
    
    xx, yy = make_meshgrid(X0, X1)

    # Polynomial kernel
    clf = SVC(kernel='linear', C=1).fit(X_train, y_train)
    Z, Z1 = find_Z(clf, xx, yy)

    # Plot
    plt.figure()
    plt.contour(xx, yy, Z1, colors = 'indigo', levels=[-1, 0, 1], linewidths=2, linestyles=['--', '-', '--'])
    plt.contourf(xx, yy, Z, cmap=cmap_light, alpha = 0.8)
    plt.scatter(X0, X1, c=y_train, marker= 'o', s=50, cmap=cmap_bold, edgecolors='k')
    plt.title("Linear Kernel with C = 1")

    # RBF Kernel
    clf = SVC(gamma=1, C = 1).fit(X_train, y_train)
    Z, Z1 = find_Z(clf, xx, yy)

    # Plot
    plt.figure()
    contours = plt.contour(xx, yy, Z1, colors = 'indigo', levels=[0], linewidths=2, linestyles='dashed')
    plt.contourf(xx, yy, Z, cmap=cmap_light, alpha = 0.8)
    plt.scatter(X0, X1, c=y_train, marker= 'o', s=50, cmap=cmap_bold, edgecolors='k')
    plt.title("RBF Kernel with C = 1, gamma = 1")

    # Polynomial kernel
    clf = SVC(kernel='poly', degree=3, gamma = 'auto').fit(X_train, y_train)
    Z, Z1 = find_Z(clf, xx, yy)
    
    # Plot
    plt.figure()
    contours = plt.contour(xx, yy, Z1, colors = 'indigo', levels=[0], linewidths=2, linestyles='dashed')
    plt.contourf(xx, yy, Z, cmap=cmap_light, alpha = 0.8)
    plt.scatter(X0, X1, c=y_train, marker= 'o', s=50, cmap=cmap_bold, edgecolors='k')
    plt.title("Polynomial Kernel with Degree = 3")

    plt.show()

answer_8b(X_train[:,0],X_train[:,1])
```

```python
# Problem 8(c)

# Set C for linear model
C_linear = [0.01, 0.1, 1, 10, 100]
scores=[]

# Calculate fitting score for test data for different values of C
for c in C_linear:
    clf = SVC(kernel='linear', C=c).fit(X_train, y_train)
    s = clf.score(X_test, y_test)
    scores.append(s)
    
print('Different rows indicate different values of C')
pd.DataFrame(data=scores,index = C_linear, columns = {'Scores'})
```

```python
# Set C and gamma for RBF model
C_RBF = [0.1, 1, 10]
gamma = [0.01, 1, 10, 100]
scores_c=[]

# Calculate fitting score for test data for different values of C and gamma
for c in C_RBF:
    scores_g = []
    for g in gamma:
        clf = SVC(gamma = g, C=c).fit(X_train, y_train)
        s = clf.score(X_test, y_test)
        scores_g.append(s)
    scores_c.append(scores_g)

print('Different rows indicate different values of C and different columns indicate different values of gamma.')
pd.DataFrame(data=scores_c,index = C_RBF, columns = gamma)
```

```python
# Set C and degree for polynomial model
C_Poly = [0.1, 1, 10]
deg = [2, 3, 4, 5]
scores_c=[]

# Calculate fitting score for test data for different degrees and values of C
for c in C_Poly:
    scores_d = []
    for d in deg:
        clf = SVC(kernel = 'poly', C = c, degree = d, gamma = 'auto').fit(X_train, y_train)
        s = clf.score(X_test, y_test)
        scores_d.append(s)
    scores_c.append(scores_d)
print('Different rows indicate different values of C and different columns indicate different degrees.')
pd.DataFrame(data=scores_c,index = C_Poly, columns = deg)
```

```python
print('In linear model, C=100 shows the best fitting score so it is the best model here.')
print('Likewise, C=10 and gamma=0.01 is the best model in RBF.')
print('In polynomial model, C=10 with degree=5 is the best model.')
```
